# Breed

```python
def distinct_col_values(dataframe, column_name):
    return dataframe[column_name].unique()

# print(distinct_col_values(puppy_info, "Breed"))
```

```python
def map_col_values(dataframe, column_name, callback):
    return dataframe[column_name].map(callback)

def breed_mappings(breed):
    if breed == "golden labrador" or breed == "golden lab":
        return "golden labrador"
    elif "rador" in breed or "lab" in breed or breed[0:2] == "la" or breed in ["lr", "l", "bl"]:
        return "labrador retriever"
    elif "german shep" in breed or breed in ["gs", "gds", "greman shepherd", "geman shepherd", "gsd", "gsp", "g.s."] or "german sh" in breed or "shep" in breed:
        return "german shepherd"
    elif breed in ["gr"] or "golden retr" in breed:
        return "golden retriever"
    elif breed in ["belgen sharpart"]:
        return "belgian shepherd"
    elif breed in ["black", "golden", "lym", "lbm", "ly m", "ged", "gdd", "noble", "nan"]:
        return "unknown"
    else:
        return breed

puppy_breeds = map_col_values(puppy_info, "Breed", lambda x: str(x).lower())

distinct_breeds = puppy_breeds.unique()
distinct_breeds.sort()

distinct_breeds_mapped = puppy_breeds.map(breed_mappings).unique()

# print(len(distinct_breeds), distinct_breeds)
# print()
print(len(distinct_breeds_mapped), distinct_breeds_mapped)
```

```python
puppy_info["Breed"] = puppy_info["Breed"].map(lambda x: breed_mappings(str(x).lower()))
puppy_info.head()
```
